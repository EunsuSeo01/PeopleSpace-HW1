```python
# Eun Su Seo (Winter), 2024/01/17
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

# Add option to pass the robot test
options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)
total_data = []

# Search for 100 pages in Books, movies & music category of Etsy
for i in range(1, 100):
    driver.get("https://www.etsy.com/c/books-movies-and-music?ref=pagination&page=" + str(i))

    # wait for loading website
    driver.implicitly_wait(20)

    # bring the page source
    html_source = driver.page_source
    # html parsing with BeautifulSoup
    soup = BeautifulSoup(html_source, 'html.parser')

    # find the elements with div tag and the class name is v2-listing-card__info
    for k in soup.find_all('div', 'v2-listing-card__info'):
        product_info = []
        # Product Name
        name = k.find('h3', 'wt-text-caption v2-listing-card__title wt-text-truncate').text
        product_info.append(name)
        print("Product Name " + name)

        # Product Price
        prices = k.find_all('span', 'currency-value')
        # If price data is found more than 1, the product is on sale
        if len(prices) > 1:
            print("Discounted price " + str(prices[0].text))
            print("Original price " + str(prices[1].text))
            product_info.append(prices[0].text)
            product_info.append(prices[1].text)
        else:
            print("Original price " + str(prices[0].text))
            product_info.append(prices[0].text)
            product_info.append(prices[0].text)

        # Product review count
        review = k.find('span', 'wt-text-caption wt-text-gray wt-display-inline-block wt-nudge-l-3 wt-pr-xs-1')
        if review is None:
            print("Product review count", 0)
            product_info.append(0)
        else:
            print("Product review count", review.text)
            product_info.append(review.text)

        # Product badge
        etsy_pick = k.find('span', 'wt-badge wt-badge--statusRecommendation wt-badge--small wt-mb-xs-1')
        if etsy_pick is None:
            print('Product have no badge')
            product_info.append('No badge')
        else:
            print("Product have " + etsy_pick.text + " badge")
            product_info.append(etsy_pick.text)

        total_data.append(product_info)

    # wait for next page
    driver.implicitly_wait(10)
```

```python
# Save data to CSV
f = open(f'final-data.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(f)
writer.writerow(['Product Name ', 'Discounted Price ', 'Original Price ', 'Review Counts ', 'Badge'])
writer.writerows(total_data)
f.close()

print("======Scraping Completed======")
print("Results saved to final-data.csv")
```
